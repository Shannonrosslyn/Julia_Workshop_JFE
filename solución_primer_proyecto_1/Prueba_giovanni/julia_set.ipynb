```julia
using Plots
```

```julia
function f(z)
return z^2-0.1+0.65im
end
```

```julia
function plot_graphics(size,map)
heatmap(size, size, map, color=cgrad([:black,:blue,:white]), 
        title="Ejemplo Mapa de Calor", xlabel="Re(z)", ylabel="Im(z)")
end 
```

```julia
function julia_algorith(z,zmax,kmax)
k=0
while (abs(z) <= zmax) && (k<kmax)
z=f(z)
k+=1
end 
value=k/kmax
return value
end 
```

```julia
size=500
min=-1.5
max=1.5
zmax=10
kmax=1000
delta=(max-min)/size
side = min:delta:max
size_range=1:size
julia_image=zeros(size,size)
for i in size_range
for j in size_range
z=side[i]+side[j]*im
julia_image[i,j]=julia_algorith(z,zmax,kmax)
end 
end
```

```julia
plot_graphics(size_range,julia_image)
```
